```python
import pandas as pd
data_set=pd.read_csv('C:/Users/lenovo/Downloads/Fast Food Sales.csv')

data_set.head()
```

```python
data_set['date'] = pd.to_datetime(data_set['date'])


data_set.head()
```

```python
data_set['date'].tail(15)
```

```python
data_set = data_set.rename(columns={'date': 'Date'})
```

```python
data_set.head()
```

```python
# Question asked from the data:
#  1. What is the overall trend  in transaction amounts  over time?
# 2. Which item name are most commonly sold?
# 3. Which item type are most commonly sold
# 4. Which transaction type is the most commonly used?
# 5. Who is the most common recipient of transactions?
# 6. When is the best-selling time?
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Question_01**

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

daily_sales = data_set.groupby('Date')['transaction_amount'].sum().reset_index()


plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='transaction_amount', data=daily_sales)
plt.title('Overall Trend in Transaction Amounts Over Time')
plt.xlabel('Date')
plt.ylabel('Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Question_02**

```python

item_sales = data_set.groupby('item_name')['quantity'].sum().reset_index().sort_values(by='quantity', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='quantity', y='item_name', data=item_sales, palette='viridis')
plt.title('Most Commonly Sold Items')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Item Name')
plt.tight_layout()
plt.show()
```

**Question_03**

```python
item_type_sales = data_set.groupby('item_type')['quantity'].sum().reset_index().sort_values(by='quantity', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x='quantity', y='item_type', data=item_type_sales, palette='rocket')
plt.title('Most Commonly Sold Item Types')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Item Type')
plt.tight_layout()
plt.show()
```

**Question_04**

```python
transaction_type_counts = data_set['transaction_type'].value_counts()

plt.figure(figsize=(7, 7))
transaction_type_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Most Commonly Used Transaction Types')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

**Question_05**

```python
recipient_counts = data_set['received_by'].value_counts()

plt.figure(figsize=(7, 7))
recipient_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Most Common Recipients of Transactions')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

**Question_06**

```python
time_sales = data_set.groupby('time_of_sale')['quantity'].sum().reset_index().sort_values(by='quantity', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x='quantity', y='time_of_sale', data=time_sales, palette='coolwarm')
plt.title('Best-Selling Times')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Time of Sale')
plt.tight_layout()
plt.show()
```
